```python
import numpy as np
import corner as DFM
# -- local -- 
import env
import util as UT
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.xmargin'] = 1
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1.5
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 1.5
mpl.rcParams['legend.frameon'] = False
%matplotlib inline
```

```python
mufasa1 = np.loadtxt(UT.dat_dir()+'MUFASA_combined.dat', skiprows=1, unpack=True, usecols=[0, 3, -1])
mufasa2 = np.loadtxt(UT.dat_dir()+'MUFASA_GALAXY_extra.txt', skiprows=17, unpack=True, usecols=[6, 8, 12])
```

```python
iscen1 = (mufasa1[2] == 1)
iscen2 = (mufasa2[2] == 1)
```

```python
fig = plt.figure()
sub = fig.add_subplot(111)
DFM.hist2d(mufasa1[0][iscen1], mufasa1[1][iscen1], color='k',
           levels=[0.68, 0.95], range=[[7., 12.], [-4., 2.]], 
           plot_density=False, ax=sub)
sub.plot([0,0],[0,0], c='k', label=r'$\mathrm{MUFASA_combined.dat}$')
DFM.hist2d(mufasa2[0][iscen2], mufasa2[1][iscen2], color='C1',
           levels=[0.68, 0.95], range=[[7., 12.], [-4., 2.]], 
           plot_density=False, ax=sub)
sub.plot([0,0],[0,0], c='C1', label=r'$\mathrm{MUFASA_GALAXY_extra.txt}$')
sub.set_xlabel('stellar mass', fontsize=20)
sub.set_ylabel('star formation rate', fontsize=20)
sub.legend(loc='lower left', prop={'size': 15})
sub.set_title(r'MUFASA $100\,\mathrm{Myr}$ SFR discrepancy', fontsize=20)
```

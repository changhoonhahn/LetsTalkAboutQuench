```python
import pickle
import numpy as np
# -- iquench -- 
from letstalkaboutquench import util as UT
```

```python
import corner as DFM
import matplotlib as mpl
import matplotlib.pyplot as plt 
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.xmargin'] = 1
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1.5
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 1.5
mpl.rcParams['legend.frameon'] = False
%matplotlib inline
```

```python
def sfs_gmmfit(name, dev_thresh=0.3): 
    '''
    '''
    f_gmm = ''.join([UT.dat_dir(), 'paper1/',
                     'gmmSFSfit.', name, '.gfcentral.mlim.dev_thresh', str(dev_thresh), '.p'])
    return pickle.load(open(f_gmm, 'rb'))
```

```python
fig = plt.figure(figsize=(16,4))
for i_sim, sim in enumerate(['illustris', 'eagle', 'mufasa', 'scsam']): 
    sub = fig.add_subplot(1,4,i_sim+1)

    for i_dev, dev in enumerate([0.2, 0.3, 0.5, 0.7, 0.9]): 
        _fsfs = sfs_gmmfit(sim+'_inst', dev_thresh=dev)
        #sub.errorbar(_fsfs._fit_logm+0.025*float(i_dev), _fsfs._fit_logsfr, yerr=_fsfs._fit_err_logssfr, fmt='.C'+str(i_dev))
        sub.fill_between(_fsfs._fit_logm, _fsfs._fit_logsfr - _fsfs._fit_err_logssfr, _fsfs._fit_logsfr + _fsfs._fit_err_logssfr, alpha=0.25, color='C'+str(i_dev))
    if i_sim == 0: 
        sub.set_ylabel('$\log\,\mathrm{SFR}$', fontsize=25)
    sub.set_xlim([8., 12.])
    sub.set_title(sim.upper(), fontsize=20)
    sub.set_yticklabels([])
```

```python
import h5py
import numpy as np 

import matplotlib.pyplot as plt
%matplotlib inline
#---local---
import env
import util as UT
```

```python
# M* from SFH catalog
f = h5py.File(UT.dat_dir()+'binsv2all1e8Msunh_z0.hdf5', 'r')
Mstar = np.log10(f['CurrentStellarMass'].value.flatten()) + 10.
```

```python
# M* from group catalog
Mstar2 = np.loadtxt(UT.dat_dir()+'Illustris1_extended_individual_galaxy_values_all1e8Msunh_z0.csv', 
                   unpack=True, skiprows=1, delimiter=',', usecols=[0])
```

```python
plt.scatter(Mstar, Mstar2 + np.log10(0.704), lw=0)
plt.plot([8.,12.], [8., 12.], c='k', lw=5, ls='--')
plt.xlim([8., 12.])
plt.xlabel('$log\,M_*$ Illustris SFH catalog', fontsize=20)
plt.ylim([8., 12.])
plt.ylabel('$log\,M_*$ Illustris group catalog', fontsize=20)
```
